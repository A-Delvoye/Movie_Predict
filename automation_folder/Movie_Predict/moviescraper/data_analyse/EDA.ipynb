```python
big_countries=df['country'][df['France_first_week']>400000].unique()

big_country=[]
for country in big_countries:
    country_list=country.split(',')
    for c in country_list:
        if c not in big_country:
            big_country.append(c)

len(big_country)
```

```python
print(big_countries)
```

```python
no_usa_gb__country=[]
for country in big_countries:
    country_list=country.split(',')
    for c in country_list:
        d = c.strip()
        if ("U.S.A." not in country  and "France" not in country):
            no_usa_gb__country.append(d)

no_usa_gb__country
```
